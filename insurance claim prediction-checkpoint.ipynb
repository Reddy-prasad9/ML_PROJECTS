## MODEL CREATION

##### the data split into x variable is independent features and y is target variable

```python
X = data.drop('target', axis=1) #
y = data['target']
```

##### BALANCING DATA THE TARGET COLUMN USING SMOTE¶

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
sm =SMOTE() 
print("unbalanced data   :  ",Counter(y))
x_s,y_s = sm.fit_resample(X,y)
print("balanced data:    :",Counter(y_s))
```

#### SPLIT TRAINING AND TESTING DATA¶

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_s,y_s, test_size= 0.3,random_state =42)
```

```python
x_train.shape
```

```python
x_test.shape
```

##### DATA fit into  standard scaler

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
X_train = sd.fit_transform(x_train)
X_test = sd.transform(x_test)
```

#### USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

model.fit(x_train,y_train)

train_pred = model.predict(x_train)

test_pred =model.predict(x_test)
```

#### TRAINING ACCURACY

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,classification_report,confusion_matrix,f1_score

train_accuracy = accuracy_score(train_pred,y_train)

# print accuracy scores
print(f"Training accuracy of Logistic regression model,{train_accuracy*100}")


# print classification report
print("Logistic regression training Classification report: \n",classification_report(train_pred,y_train)) 
```

#### TESTING ACCURACY

```python
test_accuracy = accuracy_score(test_pred,y_test)
print(f"Testing  accuracy of logistic regression model,{test_accuracy * 100}")
print("Logistic regression training Classification report: \n",classification_report(test_pred,y_test))

# print accuracy scores
```

```python
f1 = f1_score(test_pred, y_test)
print(f'F1 Score: {f1}')
```

##### Testing accuracy of logistic regression model is poor so we use hyper perameter tunning to improve accuracy and f1 score



#### HYPER PERAMETERTUNING

```python
param_grid = { 'C': [0.1, 1, 10, 100], 
              'penalty': ['l1', 'l2'],
              'solver': ['liblinear', 'saga'],
              'max_iter': [100, 200, 300] }
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, cross_val_score
cv=KFold()
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=2)
```
